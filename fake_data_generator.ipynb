```python
from faker import Faker
import pandas as pd
import csv
import random
import os 

class GenerateFakeData:
    def __init__(self, records):
        self.records = records
        self.fake = Faker(['fr_FR'])  # Correct initialization of Faker instance
        self.fake.seed_instance(4321)  # Properly seed the instance
        self.email_providers = ['gmail.com', 'yahoo.com', 'outlook.com', 'hotmail.com', 'hotmail.fr', 'outlook.fr', 'yahoo.fr', 'hotmail.de', 'outlook.de', 'yahoo.de', 'hotmail.be', 'outlook.be', 'yahoo.be', 'hotmail.ch', 'outlook.ch', 'yahoo.ch']  # List of email providers
    
    def generate_email(self, first_name, last_name):

        # Randomly truncate first name or last name
        first_name = first_name[:random.randint(1, len(first_name))]
        last_name = last_name[:random.randint(1, len(last_name))]

        # Randomly choose a special character or leave it blank
        special_char = random.choice([ '_', '.', '-'])
        # Optionally add a random number at the end
        number = str(random.randint(0, 99)) if random.choice([True, False]) else ''

        email_address = f"{first_name.lower()}{special_char}{last_name.lower()}{number}@{random.choice(self.email_providers)}"
        return email_address


    def generate_data(self):
            data = []  # Initialize an empty list to store the dictionaries
         
            
            for _ in range(self.records):
                first_name = self.fake.first_name()
                last_name = self.fake.last_name()
                email_address = self.generate_email(first_name, last_name)
                country = ['Luxembourgeois', 'Français','Belge', 'Suisse']
                month_expiration = str(random.randint(1, 12)).zfill(2)  # Ensure month is 2 digits
                year_expiration = str(random.randint(2025, 2040))
                credit_card_expire =f"{month_expiration}/{year_expiration}"
                cvv = random.randint(1,999)
                cvv = str(cvv).zfill(3)


                row = {  # Create a dictionary with the fake data
                    "id": self.fake.unique.random_number(digits=10, fix_len=True),
                    "name": f"{first_name} {last_name}",
                    "email_address":  email_address,
                    "address": self.fake.address().replace("\n", " "),
                    "phone_number": self.fake.phone_number(),
                    "company": self.fake.company(),
                    "job_title": self.fake.job(),
                    "date_of_birth": self.fake.date_of_birth(minimum_age=18, maximum_age=67).strftime('%Y-%m-%d'),
                    "country_of_birth": random.choice(country),
                    "member_since": self.fake.date_time_between(start_date="-2y", end_date="now").strftime('%Y-%m-%d %H:%M:%S'),
                    'cred_card_provider': self.fake.credit_card_provider(),
                    'credit_card_number': self.fake.credit_card_number(),
                    'credit_card_expire': credit_card_expire ,
                    'credit_card_cvv': cvv,
                }
                data.append(row)  # Append the dictionary to the list
            
            return data  # Return the list of dictionaries



generator = GenerateFakeData(1000)


# Convert the list of dictionaries to a Pandas DataFrame
fake_data_df = pd.DataFrame(generator.generate_data())

downloads_folder = os.path.expanduser("~/Downloads")

# Full path to save the CSV file
csv_file_path = os.path.join(downloads_folder, "fake_data.csv")

# Save the DataFrame to a CSV file in the Downloads folder
fake_data_df.to_csv(csv_file_path, index=False, quoting=csv.QUOTE_ALL, encoding="utf-8-sig")

print(f"Fake data generated and saved to {csv_file_path}")
```
